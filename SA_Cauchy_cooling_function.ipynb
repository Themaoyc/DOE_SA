### Info matrix and DB

```python
def InfoMNL(X, b, cset):
    r, c = X.shape
    nchoices = r // cset
    info = np.zeros((c, c))
    for i in range(cset):
        xrow = X[i * nchoices : i * nchoices + nchoices, :]
        u = np.dot(xrow, b)
        exp_u = np.exp(u)
        p = exp_u / np.sum(exp_u)
        Pdiag = np.diag(p)
        info += np.dot(xrow.T, (Pdiag - np.outer(p, p)).dot(xrow))
        
    return info

def calcObjFun_Bayesian(xmat, pts, wts, cset):
    objfun = 0
    for idx in range(len(wts)):
        b = pts[:,idx]
        info = InfoMNL(xmat, b, cset)
        det_info = np.linalg.det(info)
        if det_info <= 0:
            objfun = -100000
        else:
            objfun += np.log(det_info) * wts[idx]
            
    return objfun
```

### Random starting

```python
def random_start(cset, n_alt, nlevels):
    data = []
    for level in nlevels:
        data.append(np.random.choice(range(1, level+1), size=(cset*n_alt)))
        result = np.stack(data, axis=1)
    return result
```

### Effective coding

```python
def eff_code(X, nlevels):
    nf = len(nlevels)
    df = sum(nlevels) - nf
    xrow = np.zeros(df)
    startidx = 0
    for i in range(nf):
        nl = nlevels[i]
        xtmp = np.zeros(nl-1)
        if X[i] < nl:
            xtmp[int(X[i])-1] = 1
        else:
            xtmp[:nl-1] = -1
        xrow[startidx:startidx+nl-1] = xtmp
        startidx += nl - 1
    return xrow
```

### Random update

```python
def random_modify_element(X, nlevels):
    row_idx = np.random.randint(0, X.shape[0])
    col_idx = np.random.randint(0, X.shape[1])
    current_X = X.copy()
    current_value = current_X[row_idx, col_idx]
    all_possible_values = list(range(1, nlevels[col_idx] + 1))
    all_possible_values.remove(current_value)
    new_value = np.random.choice(all_possible_values)
    current_X[row_idx, col_idx] = new_value
    return current_X
```

```python
def Bayesian_D_optimal_SA(cset,n_alt,nlevels, pts, wts,initial_temp, max_iter):
    start_time = time.time()
    results_list = []
    prob_df = pd.DataFrame()
    current_X = random_start(cset, n_alt, nlevels)
    current_X_code= np.apply_along_axis(lambda row: eff_code(row, nlevels), axis=1, arr=current_X)
    current_D = calcObjFun_Bayesian(current_X_code, pts, wts, cset)
    temp = initial_temp
    t=0
    while t <= max_iter:
        new_X = random_modify_element(current_X, nlevels)
        new_X_code= np.apply_along_axis(lambda row: eff_code(row, nlevels), axis=1, arr=new_X)
        new_D = calcObjFun_Bayesian(new_X_code, pts, wts, cset)
        prob = np.exp((new_D-current_D) / temp)
        if new_D >= current_D:
            current_X = new_X
            current_D = new_D
            prob = None
        elif np.random.rand() < prob:
            current_X = new_X
            current_D = new_D
        current_time = time.time()-start_time
        print(f"t: {t}, DB: {current_D}, time: {current_time},prob:{prob}")
        D_df = pd.DataFrame({'D_value': [current_D]})
        results_list.append(current_D)
        t +=1
        temp = initial_temp/t
    result = np.array(results_list)
    end_time = time.time()
    total_time = end_time - start_time
    return current_X, current_D,result,total_time 
```
